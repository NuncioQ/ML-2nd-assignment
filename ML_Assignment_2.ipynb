```python

import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

#in this case the data is loaded as training set and test set containing both x and y values
(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)


def normalize_img(image, label):
    """Normalizes images"""
    return tf.cast(image, tf.float32) / 255.0, label

ds_train = ds_train.map(normalize_img)

#as done before, for each instance the values of n and act_func will change
n = 5
act_func = 'relu'

classifier = tf.keras.models.Sequential()  #whole model is set up as in part B of the exercise
classifier.add(tf.keras.layers.Flatten(input_shape = (28, 28, 1)))
for i in range(n):
  classifier.add(tf.keras.layers.Dense(32, activation=act_func))
classifier.add(tf.keras.layers.Dense(10, activation='softmax'))
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
loss = tf.keras.losses.CategoricalCrossentropy()


max_gradient_relu = {} #those variables will help keep track of the gradients and plot them later
max_grad_idx = []

epochs = 3   #in this loop the model is trained somewhat manually to extract the gradient values
for epoch in range(epochs):
    for step, (x_batch_train, y_batch_train) in enumerate(ds_train.take(50).batch(64)): #enumerate is used to provide an index for each iteration
        x_batch_train = tf.cast(x_batch_train, dtype=tf.float32)  #ds_train is a tensorflow object, the values of x and y must be extracted
        y_batch_train = tf.keras.utils.to_categorical(y_batch_train, 10)

        with tf.GradientTape() as tape:
          logits = classifier(x_batch_train, training=True)   #x_batch_train values are passed through the classifier
          loss_value = loss(y_batch_train, logits)  #computing loss value

        grads = tape.gradient(loss_value, classifier.trainable_weights)  #gradients of the loss_value with respect to the trainable_weights of the classifier model are computed using the gradient tape
        opt.apply_gradients(zip(grads, classifier.trainable_weights)) #computed gradients are then applied to the trainable weights using the chosen optimizer

        with tf.GradientTape(persistent=True) as tape:  #Another gradient tape is created, this time with the persistent argument set to True which means that the tape can be used multiple times to compute gradients
          tape.watch(x_batch_train)  #x_batch_train is then watched by the gradient tape, and the output of each layer in the classifier model is computed and appended to the outputs list
          x = classifier.layers[0](x_batch_train)
          outputs = []
          for layer in classifier.layers[1:]:
              x = layer(x)
              outputs.append(x)
          outputs.pop(n)  #The output of the last layer in the classifier model is then removed from the outputs list
        for idx, output in enumerate(outputs):
           grad = tf.math.abs(tape.gradient(output, x_batch_train))
           max_grad_idx.append(float('{}'.format(tf.reduce_max(grad))))
        max_gradient_relu[epoch] = max_grad_idx
        max_grad_idx=[]


#the same code is used in all instances by changing the values of n and act_func. The dataset is loaded in each instance to ensure the data hasnt been altered

(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)
ds_train = ds_train.map(normalize_img)

act_func = 'tanh'

classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Flatten(input_shape = (28, 28, 1)))
for i in range(n):
  classifier.add(tf.keras.layers.Dense(32, activation=act_func))
classifier.add(tf.keras.layers.Dense(10, activation='softmax'))
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
loss = tf.keras.losses.CategoricalCrossentropy()


max_gradient_tanh = {}
max_grad_idx = []

epochs = 3
for epoch in range(epochs):
    for step, (x_batch_train, y_batch_train) in enumerate(ds_train.take(50).batch(64)):
        x_batch_train = tf.cast(x_batch_train, dtype=tf.float32)
        y_batch_train = tf.keras.utils.to_categorical(y_batch_train, 10)

        with tf.GradientTape() as tape:
          logits = classifier(x_batch_train, training=True)
          loss_value = loss(y_batch_train, logits)

        grads = tape.gradient(loss_value, classifier.trainable_weights)
        opt.apply_gradients(zip(grads, classifier.trainable_weights))

        with tf.GradientTape(persistent=True) as tape:
          tape.watch(x_batch_train)
          x = classifier.layers[0](x_batch_train)
          outputs = []
          for layer in classifier.layers[1:]:
              x = layer(x)
              outputs.append(x)
          outputs.pop(n)
        for idx, output in enumerate(outputs):
           grad = tf.math.abs(tape.gradient(output, x_batch_train))
           max_grad_idx.append(float('{}'.format(tf.reduce_max(grad))))
        max_gradient_tanh[epoch] = max_grad_idx
        max_grad_idx=[]





(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)
ds_train = ds_train.map(normalize_img)

act_func = 'sigmoid'

classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Flatten(input_shape = (28, 28, 1)))
for i in range(n):
  classifier.add(tf.keras.layers.Dense(32, activation=act_func))
classifier.add(tf.keras.layers.Dense(10, activation='softmax'))
opt = tf.keras.optimizers.SGD(learning_rate=0.01)
loss = tf.keras.losses.CategoricalCrossentropy()


max_gradient_sigmoid = {}
max_grad_idx = []

epochs = 3
for epoch in range(epochs):
    for step, (x_batch_train, y_batch_train) in enumerate(ds_train.take(50).batch(64)):
        x_batch_train = tf.cast(x_batch_train, dtype=tf.float32)
        y_batch_train = tf.keras.utils.to_categorical(y_batch_train, 10)

        with tf.GradientTape() as tape:
          logits = classifier(x_batch_train, training=True)
          loss_value = loss(y_batch_train, logits)

        grads = tape.gradient(loss_value, classifier.trainable_weights)
        opt.apply_gradients(zip(grads, classifier.trainable_weights))

        with tf.GradientTape(persistent=True) as tape:
          tape.watch(x_batch_train)
          x = classifier.layers[0](x_batch_train)
          outputs = []
          for layer in classifier.layers[1:]:
              x = layer(x)
              outputs.append(x)
          outputs.pop(n)
        for idx, output in enumerate(outputs):
           grad = tf.math.abs(tape.gradient(output, x_batch_train))
           max_grad_idx.append(float('{}'.format(tf.reduce_max(grad))))
        max_gradient_sigmoid[epoch] = max_grad_idx
        max_grad_idx=[]

layers_num=[]
for i in range(n):
    layers_num.append(i+1)

plt.xlabel('Layer #')
plt.ylabel('Max gradient')

plt.plot(layers_num, max_gradient_relu[0], 'b', label = 'Relu')
plt.plot(layers_num, max_gradient_tanh[0] , 'y' , label = 'tanh')
plt.plot(layers_num, max_gradient_sigmoid[0], 'r', label = 'Sigmoid')

plt.plot(layers_num, max_gradient_relu[1], 'b')
plt.plot(layers_num, max_gradient_tanh[1], 'y')
plt.plot(layers_num, max_gradient_sigmoid[1], 'r')

plt.plot(layers_num, max_gradient_relu[2], 'b')
plt.plot(layers_num, max_gradient_tanh[2], 'y')
plt.plot(layers_num, max_gradient_sigmoid[2], 'r')

plt.legend()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from tensorflow import keras
from keras import datasets, layers, models, optimizers
from keras.utils import plot_model, to_categorical
from keras.models import Model, load_model
from keras.layers import BatchNormalization, Conv2D, Dense, Dropout, Flatten, Input, MaxPooling2D, Reshape
from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

The **MNIST dataset** is a collection of 70,000 grayscale, with a size of 28x28 pixels images of handwritten digits ranging from 0 to 9. The dataset is split into 60,000 training images and 10,000 test images.

```python
# Loading the dataset
(train_im, train_lbl), (test_im, test_lbl)= datasets.mnist.load_data()

# Displaying the first 9 images in training data
plt.figure(figsize=[3,3])
for i in range(9):
  plt.subplot(3 ,3 ,1 + i)
  plt.imshow(train_im[i], cmap='gray')
  plt.axis('off')
plt.show()
```

**Data Preprocessing**


We will the image and label data in a format that can be easily fed into the neural network. Here are the steps being performed:

* Reshaping the 2D images into 4D tensors with a single channel. The -1 in the reshape function means that the size of that dimension is inferred from the size of the other dimensions and the total number of elements in the array (number of images in the dataset).

* Converting the pixel values from integers to floats using the astype function.

* Scaling the pixel values to the range [0,1]. The pixel values range from 0 to 255, as seen above, so we will divide each pixel value with 255. This step makes it easier for the model to learn.

* Finding the unique labels in the training set using the np.unique function. This is important because it tells us how many output classes the model needs to predict (which is expected to be 10 - digits 0 - 9).

```python
# Reshaping into a single sample with 1 channel
train_images = train_im.reshape(-1, 28, 28, 1)
test_images = test_im.reshape(-1, 28, 28, 1)

# Converting from integers to floats
train_images = train_images.astype('float32')
test_images = test_images.astype('float32')

# Scaling the pixel values to the [0,1] range
train_images = train_images / 255.0
test_images = test_images / 255.0

# Finding the unique numbers from the train labels
classes = np.unique(train_lbl)
numofclasses = len(classes)
print('Total number of outputs : ', numofclasses)
print('Output classes : ', classes)
```

Now we will convert the class labels into one-hot encoding vectors. In one-hot encoding, you convert the categorical data into a vector of numbers. The reason why you convert the categorical data in one hot encoding is that machine learning algorithms cannot work with categorical data directly. The resulting vector consists of all zeros except for the class that it represents, and for that, it is 1.

```python
# Changing the labels from categorical to one-hot encoding
train_labels_1h = to_categorical(train_lbl)
test_labels_1h = to_categorical(test_lbl)

# Displaying the change for category label using one-hot encoding
print('Original label:', train_lbl[0])
print('After conversion to one-hot:', train_labels_1h[0])
```

For the model to generalize well, we will split the training data into two parts, one designed for training and another one for validation. So, we will train the model on 80% of the training data and validate it on 20% of the remaining training data.

```python
# Splitting the dataset into a training set and a validation set.
train_images, valid_images, train_labels, valid_labels = train_test_split(train_images, train_labels_1h, test_size=0.2)
```

**Compiling and training the model**

We will first add a first convolutional layer. On the first layer our network will learn 32 convolutional filters, each of which with a 3x3 size. The activation is relu, which is a simple way of introducing non-linearity. On the second layer of our network we have again 32 3x3 convolutional filters, relu as the activation function, and so on. On the fifth, tenth and fifteenth layer we also have a max-pooling operation with pool size 2x2. The last layer is a Dense layer that has a softmax activation function with 10 units, which is needed for this multi-class classification problem.

```python
inputs = Input(shape=(28, 28, 1))
conv1 = Conv2D(32, (3,3), activation='relu', padding='same')(inputs)
mpool1 = MaxPooling2D(pool_size=(2,2))(conv1)
conv2 = Conv2D(64, (3,3), activation='relu', padding='same')(mpool1)
mpool2 = MaxPooling2D(pool_size=(2,2))(conv2)

f1 = layers.Flatten()(mpool2)
d1 = layers.Dense(64 , activation='relu')(f1)
d2 = layers.Dense(128, activation='relu')(d1)
outputs = layers.Dense(10, activation='softmax')(d2)

model = Model(inputs=inputs, outputs=outputs)
model.summary()
model.compile(loss='categorical_crossentropy', optimizer = optimizers.Adam(), metrics=['accuracy'])

model.fit(train_images, train_labels, batch_size=64, epochs=5, validation_data=(valid_images, valid_labels))
```

```python
# Creating a utility function that can be used to visualize the predicted probabilities of a single input sample using our network
# The function below takes as input a sample x and a predictor model

def predict(x, predictor):
  activations = predictor.predict(x.reshape(dim))    # Reshaping the input sample to the shape expected by the predictor model
  predictions = np.argmax(activations)    # The predicted label for the input sample is the class with the highest predicted probability
  print('Network prediction: ', predictions)
  plt.bar(range(10), activations[0], width=0.8, alpha=0.8)
  plt.show()

# In this part of the exercise (2A) we have 32x32 images, so dim is (32, 32, 3)
dim = (1, 28, 28, 1)

predict(test_images[0], model)
print('Actual image: ')
plt.imshow(test_images[0].reshape((28,28)), cmap='gray')
plt.show()
```

**Constructing adversarial examples**

We will use another network to add noise to the images and create adversarial examples that will be able to consistently fool the network with high probability.

To achieve this, we will use an adversarial model that will modify the original images to create the adversarial images. The adversarial model will be trained to minimize the mean squared error between the original and adversarial images. We will then combine the adversarial model with our original model to create a whole model that can classify the adversarial images. We will use categorical cross-entropy as the loss function for the whole model.


```python
model.trainable = False   # Freezing the weights of the model so that they will not be updated during training of the adversarial model

input = Input(shape=(1,))   # Input layer that takes a 1D tensor as input
hidden1 = Dense(256, activation='relu')(input)
hidden2 = Dense(28*28, activation='sigmoid')(hidden1)
output = Reshape((28,28))(hidden2)   # The output becomes a 28x28 array

# The advmodel contributes by changing the adversarial image in a way that it will eventually look a lot alike the original image (This is why we use mean squared error)

advmodel = Model(input, output, name='advmodel')
advmodel.compile(loss='mean_squared_error', optimizer=optimizers.Adam(learning_rate=0.01), metrics=[keras.metrics.RootMeanSquaredError()])
advmodel.summary()

# Creating a new model, which is is the adversary model we built above combined with our model. So basically we feed our model with the output of the adversary model
# The wholemodel contributes by changing the adversarial image in a way that our model will classify it as a different digit (This is why we use categorical crossentropy)
wholemodel = Model(input, model(output), name='wholemodel')
wholemodel.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=0.009), metrics=['accuracy'])

wholemodel.summary()
```
